Q9. **Convert the `acceptance_rate` column from object to float by removing the `%` symbol.**

```python
#step 1: Use lambda function within apply method to change type
prop_df['acceptance_rate'].apply(lambda x: type(x))
```

```python
#step 2: To convert from obj to float create a function 
def conv_acc_rate(x):
    if pd.isnull(x):
        return np.nan
    else:
        return int(x[:-1])/100
```

```python
#step 3: Give the assigned function name in place of type(x) in lambda function 
prop_df['acceptance_rate'].apply(lambda x: conv_acc_rate(x))
```

```python
prop_df['acceptance_rate'].str.slice(stop=-1).astype(float)/100
```

```python
prop_df['acceptance_rate'].str.replace('%', '').astype(float).div(100)
```
